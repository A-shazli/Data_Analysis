### Research Question 1 (Which year has the highest number of released movies?)

```python
# group by year and finding the number of movies based on id
data=df.groupby('release_year').count()['id']
#using data.tail() is more usfull to inspect the number of movies based on the graph
print(data.tail())

#make groups of the data according to their release year and count the total number of movies in each year.
#arange was used to set the lengh of the x axis
data.plot(xticks = np.arange(1960,2016,5))

#set the figure size and labels
sns.set(rc={'figure.figsize':(10,5)})
plt.title("Num of movies per year",fontsize = 15) #only increasing the title font size to be easy on the eye
plt.xlabel('Release year')
plt.ylabel('Number Of Movies')
#set the style sheet
sns.set_style("darkgrid")
```

#### After inspecting the data and as seen in the graph it can be seen that most movies were released in 2014


### Research Question 2  (Highest and lowest budget movies?)

```python
def min_max(x):
    """This function takes in a parameter that is a column you want to inspect about and retuns its max and min"""
    

    #use the function 'idmin' to find the index of lowest budget movie.
    min_index = zero_df[x].idxmin()
    #use the function 'idmax' to find the index of Highest budget movie.
    high_index = zero_df[x].idxmax()
    high = pd.DataFrame(zero_df.loc[high_index,:])
    low = pd.DataFrame(zero_df.loc[min_index,:])
    
    #print the movie with high and low budget
    print("Movie Which Has Highest "+ x + " : ",zero_df['original_title'][high_index])
    print("Movie Which Has Lowest "+ x + "  : ",zero_df['original_title'][min_index])
    return pd.concat([high,low],axis = 1)

```

```python
min_max('budget')
```

##### As seen, the first column presents info about the highest budget movie and the second one presents info about the counter part

##### Here we used the zero_df data frame since all zero budget and revenue value rows were removed for better and more meaningful results

### Note that the previous min_max function can be used to answer many more questions like which movie had the longest runtime and which was the shortest, also which movie was most popular and which was least popular and so on and so forth also profit related inquiries as follows in question 3

### Research Question 3  (Highest and lowest profite achieved movies?)

```python
#here we used the zero_df data frame that has only non zero values for revenue and budget
zero_df['Profit'] = zero_df['revenue'] - zero_df['budget']
min_max('Profit')
```

**As seen above, the previously written functon is used but this time the profit is presented to indicate the gain (highest movie) to the right and loss (lowest movie) to the left**

### Research Question 4  (Actor appearing in most movies?)

```python
#make a function will will split the string and return the count of each actor.
def count_cast(x):
    '''This functon take in a column name and breaks that column at locations where the pipes are and returns 
    counts for each value separated by the pipe'''
    #concatenate all the rows of the cast.
    data_plot = df[x].str.cat(sep = '|')
    data = pd.Series(data_plot.split('|'))
    #counts each of the cast and return.
    info = data.value_counts(ascending=False)
    return info
```

```python
#call the 'count_cast' function it wiil count number of movies done by each actor.
%matplotlib qt
count_actor_flims = count_cast('cast')

#plot the bar plot.
count_actor_flims.iloc[:20].plot.bar(figsize=(21,17),fontsize=15, color=(0.2, 0, 0.6, 0.6))

#setup the title and the labels of the plot.
plt.title("Actor Frequency chart",fontsize=20)
plt.xticks(rotation = 70)
plt.xlabel('Actor')
plt.ylabel("Number Of Movies",fontsize=20)
```

**As can be seen in the above chart, the actor casting in most movies was "Robert De Niro" appearing in 70 movies**

### Research Question 5  (What Kind Of Properties Are Associated With Movies With High Revenue?)

```python
# To answer this question we different columns agianst revenue and observe the correlation (a line of best fit is seen)
# Also notice that we used the zero_df since it contains more realistic budget and revenue values

fig, axes = plt.subplots(2,2,figsize = (16,6))
fig.suptitle("Revenue Vs (Budget,Popularity,Vote Average,Runtime)",fontsize=14)
#info.plot(x = ['revenue'],y = ['budget'],kind = "scatter",ax = axes[0][0])
sns.regplot(x=zero_df['revenue'], y=zero_df['budget'],color='c',ax=axes[0][0])
sns.regplot(x=zero_df['revenue'], y=zero_df['popularity'],color='c',ax=axes[0][1])
sns.regplot(x=zero_df['revenue'], y=zero_df['vote_average'],color='c',ax=axes[1][0])
sns.regplot(x=zero_df['revenue'], y=zero_df['runtime'],color='c',ax=axes[1][1])

sns.set_style("whitegrid")
```

**As can be seen above, there is a high correlation between budget and revenue**
